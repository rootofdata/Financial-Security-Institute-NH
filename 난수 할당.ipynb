```python
from random import *
def gab(title):
    lis1=list()
    b=list()
    for j in range(len(df[title])):
        if j%6==0:
            num=df[title][j]
            b.append(num)
        else:
            for i in range(500): #난수 개수 :500으로 지정 -> 더 늘려줘도 된다.
                num=uniform(df[title][j]*0.99,df[title][j]*1.01)
                lis1.append(num)
            a=np.mean(lis1)
            b.append(a)
    return(b)
```
